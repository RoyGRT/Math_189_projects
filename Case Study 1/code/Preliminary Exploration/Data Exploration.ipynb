```python
import pandas as pd
import matplotlib.pyplot as pyplot
import numpy as np
```

```python
babies3 = pd.read_csv("babies23.csv")
```

```python
babies3 = babies3.replace(999, np.nan)
```

```python
babies_no_smoke = babies3[babies3["smoke"] == 0]
babies_smoke = babies3[babies3["smoke"] == 1]
babies_smoke
```

```python
# plain smoking vs non-smoking
pyplot.hist(babies_no_smoke["wt"].sample(n=100, replace = True), label = "no_smoke", alpha = 0.5, color = "blue")
pyplot.hist(babies_smoke["wt"].sample(n=100, replace = True), label = "smoke", alpha = 0.5, color = "red")
pyplot.title(label = "Sample weights of babies of smoking moms vs non smoking moms")
pyplot.legend(loc = "upper right")
pyplot.ylabel('Amount of babies')
pyplot.xlabel("Birthweight in Ounces")
```

```python
# heavy smoking categories
heavy = babies3["number"] > 4
current = babies3["smoke"] == 1
heavy_current_smoke = babies3[heavy& current ]
heavy_current_smoke
```

```python
#light smoking category
light = babies_smoke["number"] < 4
light_current_smoke = babies_smoke[light] 
light_current_smoke
```

```python
pyplot.hist(light_current_smoke["wt"].sample(n=100, replace = True), label = "light", alpha = 0.5, color = "Yellow")
pyplot.hist(heavy_current_smoke["wt"].sample(n=100, replace = True), label = "heavy", alpha = 0.5, color = "Purple")
pyplot.title(label = "Sample weights of babies from heavy smokers vs light smokers")
pyplot.legend(loc = "upper right")
pyplot.ylabel('Amount of babies')
pyplot.xlabel("Birthweight in Ounces")
```

```python
pyplot.scatter(babies3["wt"].sample(n=200, replace = True), babies3["wt.1"].sample(n=200, replace = True) )
pyplot.ylabel("Mother's weight before pregnancy")
pyplot.xlabel("Child's Birthweight in Ounces")
pyplot.title("Sample of mother's pre-pregnancy weight shown with their child's birthweight")
```

```python
currently = babies3[babies3["time"] == 1]
during = babies3[babies3["time"] == 2]
one_yr = babies3[babies3["time"] == 3]
```

```python
pyplot.hist([currently["wt"].sample(n=100, replace = True),during["wt"].sample(n=100, replace = True),one_yr["wt"].sample(n=100, replace = True)], label = ['Still_Smoking', "Quit Midway", "Quit 1 year ago"])
pyplot.legend(loc ="upper right")
pyplot.title("Comparison of Birthweights based on when the mother quit smoking")
pyplot.ylabel('Amount of babies')
pyplot.xlabel("Birthweight in Ounces")
```
